```python
# Nhập các thư viện cần thiết trong quá trình thực hiện
import vnstock as vnst
import pandas as pd
import numpy as np
```

```python
# Danh sách các công ty niêm yết trên sàn 
List = vnst.listing_companies()
List
```

```python
# Thiết lập PE trung bình ngành
params2 = {
            "exchangeName": "HOSE,HNX,UPCOM"
        }
coOn3Exchange = vnst.stock_screening_insights(params=params2, size=1700, drop_lang='vi')
average_PE_byIndustry = coOn3Exchange.groupby('industryName.en').agg(
    total_pe = ('pe', 'sum'),
    avg_pe = ('pe', 'mean')
)
average_PE_byIndustry
```

```python
# Tạo danh sách công ty thỏa mãn các tiêu chuẩn cơ bản sau:
# Sàn giao dịch: HOSE, HXN, UPCOM
# Vốn hóa thị trường: trên 100 tỷ
# ROE trên 20%
params = {
            "exchangeName": "HOSE,HNX,UPCOM",
            "marketCap": (100,99999999),
            "roe": (0.2, 99)
        }
Stocks1 = vnst.stock_screening_insights (params, size=1700, drop_lang='vi')
Stocks1
```

```python
# Danh sách các chỉ số tài chính cần quan tâm
Stocks1_Index = Stocks1.columns
Stocks1_Index
```

```python
# Tăng trưởng doanh thu/ lợi nhuận so với quý trước và tăng trưởng EPS trong năm dương
Stocks2 = Stocks1[(Stocks1['lastQuarterRevenueGrowth'] > 0) & (Stocks1['epsGrowth1Year'] > 0)]
Stocks2.reset_index(drop = True, inplace = True)
Stocks2
```

```python
# Merge DataFrame2 với DataFrame1 dựa trên cột 'ticker'
df_merged = pd.merge(Stocks2, List[['ticker', 'industry', 'organName']], on='ticker', how='left')
# Thuộc tính của dataframe 
df_merged.columns
# Dataframe cuối và các thuộc tính cần quan tâm
Stocks_result = df_merged[['ticker', 'marketCap', 'roe', 'stockRating', 'businessModel','financialHealth', 'alpha','beta','pe', 'pb', 'evEbitda', 'revenueGrowth1Year','epsGrowth1Year','foreignVolumePercent','foreignBuySell20Session', 'eps', 'lastQuarterRevenueGrowth','industry', 'industryName.en','organName']]
Stocks_result
```

```python
# Tạo ra một cột mới là trung bình ngành của businessModel và financialHealth
Stocks_result['mean_health_industry'] = Stocks_result.groupby('industryName.en')[['businessModel', 'financialHealth']].transform('mean').mean(axis=1)
Stocks_result['mean_health'] = Stocks_result[['businessModel', 'financialHealth']].mean(axis=1)
Stocks_result
```

```python
# Gộp PE trung bình ngành thành một cột
industry_pe_avg = Stocks_result.groupby('industryName.en')['pe'].mean()
Stocks_result['Industry_PE_Avg'] = Stocks_result['industryName.en'].map(industry_pe_avg)
Stocks_result
```

```python
# Lọc cổ phiếu theo điều kiện sức khỏe trung bình lớn hơn hoặc bằng sức khỏe trung bình ngành và các cổ phiếu có pe bé hơn mức pe trung bình ngành
Stocks_result_ratio = Stocks_result[(Stocks_result['mean_health'] >= Stocks_result['mean_health_industry']) & (Stocks_result['pe'] <= Stocks_result['Industry_PE_Avg'])]
# Stocks_result_ratio = Stocks_result_ratio.set_index('ticker')
Stocks_result_ratio.reset_index(drop=True, inplace=True)
Stocks_result_ratio
```

```python
# Khối lượng trung bình 20 phiên gần nhất của từng cổ phiếu: Volume trung bình >= 1 triệu
# Tạo ra một dataframe theo volume rỗng
df_volume = pd.DataFrame(columns=['ticker','volume_average', 'numSession'])

indicesToRemove_Vol = []
for index, co in Stocks_result_ratio.iterrows():
    stockHisData = vnst.stock_historical_data(symbol=co[0], start_date="2024-01-23", end_date="2024-02-26", resolution="1D", type="stock", beautify=True, decor=False, source='DNSE')
    print("stock ", index, "th")
    print("length", len(stockHisData))
    
    mean_volume = stockHisData['volume'].mean()
    df_volume.loc[len(df_volume)] = {'ticker' : co[0], 'volume_average': mean_volume , 'numSession': len(stockHisData) }
    print("mean volume of last ", len(stockHisData)," session", mean_volume)
```

```python
# Merge DataFrame2 với DataFrame1 dựa trên cột 'ticker'
df_merged2 = pd.merge(Stocks_result_ratio, df_volume[['volume_average','ticker']], on='ticker', how='left')

# Kết quả sẽ là df_merged chứa các chỉ số tài chính của 100 công ty cùng với cột industry và companyname từ DataFrame1
df_merged2
```

```python
# lọc ra những cổ phiếu có mức volume trung bình của 20 phiên gần nhất lớn hơn 100000
df_merged2.drop(df_merged2[df_merged2['volume_average'] <= 100000].index, inplace = True)
df_merged2.reset_index(drop=True, inplace=True)
df_merged2
```

```python
# Dataframe cuối cùng là datafame được sắp xếp stockRating từ lớn đến nhỏ
df_merged_final = df_merged2.sort_values(by='stockRating', ascending=False)
df_merged_final
```

```python
# Lọc những dòng nếu bị NAN thì sẽ tự động drop
def drop_nan_rows(df):
    nan_rows = df[df['stockRating'].isna() | df['businessModel'].isna() | df['volume_average'].isna()]
    df_cleaned = df.dropna(subset=['stockRating', 'businessModel', 'volume_average'])
    return df_cleaned

# Sử dụng hàm drop_nan_rows trên DataFrame df_merged_final
df_merged_final = drop_nan_rows(df_merged_final)
# set index cho dataframe theo cột ticker
df_merged_final = df_merged_final.set_index('ticker')
# Dataframe cuối cùng
df_merged_final
```
