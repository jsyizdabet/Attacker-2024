```python
import vnstock as vnst
```

```python
# filter using vnstock
# Sàn giao dịch: HOSE, HXN, UPCOM
# Vốn hóa thị trường: trên 100 tỷ
# ROE trên 20%
params = {
            "exchangeName": "HOSE,HNX,UPCOM",
            "marketCap": (100,99999999),
            "roe": (0.2, 99)
        }
filter1 = vnst.stock_screening_insights (params, size=1700, drop_lang='vi')
```

```python
filter1.shape
# filter1.columns
```

```python
# Tăng trưởng doanh thu/ lợi nhuận so với quý trước
filter2 = filter1[filter1['lastQuarterRevenueGrowth'] > 0]
filter2.reset_index(drop = True, inplace = True)
```

```python
# Tăng trưởng EPS so với quý trước: EPSchange > 0
indicesToRemove = []
for index, co in filter2.iterrows():
    print(co[0])
    ptTaiChinh = vnst.financial_ratio(co[0], 'quarterly', True)
    if not ('Q3-2023' in ptTaiChinh.columns):
        indicesToRemove.append(index)
    else:
        if ptTaiChinh['Q3-2023']['epsChange'] <= 0:
            indicesToRemove.append(index)
            print(ptTaiChinh['Q3-2023']['epsChange'])
```

```python
indicesToRemove
```

```python
filter2.drop(filter1.index[indicesToRemove], inplace = True)
```

```python
filter2.reset_index(drop = True, inplace = True)
filter2
```

```python
# Khối lượng trung bình 20 phiên gần nhất của từng cổ phiếu: Volume trung bình > 100 ngàn
indicesToRemove_Vol = []
for index, co in filter2.iterrows():
    stockHisData = vnst.stock_historical_data(symbol=co[0], start_date="2018-06-04", end_date="2018-07-01", resolution="1D", type="stock", beautify=True, decor=False, source='DNSE')
    print("stock ", index, "th")
    print("length", len(stockHisData))
    
    # Tính tổng volume của 20 phiên gần nhất
    totalLast20SessVolume = 0
    for i in range(len(stockHisData)):
#         print(stockHisData.iloc[i][5])
        totalLast20SessVolume += stockHisData.iloc[i]['volume']
    print("total last 20 session", totalLast20SessVolume)
    if (totalLast20SessVolume/20) < 100000:
        indicesToRemove_Vol.append(index)
```

```python
indicesToRemove_Vol
```

```python
filter2.drop(filter2.index[indicesToRemove_Vol], inplace = True)
```

```python
filter2.reset_index(drop = True, inplace = True)
```
