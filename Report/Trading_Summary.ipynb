## Năm 2019 - 2020

### Lọc cổ phiếu

```python
trading_year = 2019
```

##### Tải file dữ liệu thô

* Dữ liệu thô được nhóm sử dụng vnstock để tải và lưu local để tiện cho việc tải dữ liệu cho bộ lọc được nhanh chóng hơn
* Chi tiết tại file .inpynb trong thư mục Data

```python
file_name = f'data_Q3-{trading_year-1}-mergedIndustry.csv'
current_directory = os.getcwd()
file_path = os.path.join(current_directory, 'Data', file_name)
df = pd.read_csv(file_path)
df.drop('Unnamed: 0', axis=1, inplace=True)
df.drop_duplicates(subset='ticker', keep='first', inplace=True)
df_dropColumns = df[['ticker', 'quarter', 'year', 'priceToEarning', 'priceToBook', 'roe',
    'roa', 'earningPerShare', 'bookValuePerShare','epsChange','bookValuePerShareChange', 'industry', 'organName', 'equityOnTotalAsset']]
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=pd.errors.SettingWithCopyWarning)
    df_dropColumns['pe_avg'] = 0
df_dropColumns.reset_index(drop=True, inplace=True)
df_dropColumns
```

##### Lọc theo giá trị PE < PE trung bình ngành

```python
# Tạo dataframe chứa dữ liệu P/E trung bình theo ngành
df_pe_byIndustry = df_dropColumns.groupby('industry').agg(
    avg_pe = ('priceToEarning', 'mean')
)
with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=FutureWarning)
    # Thêm giá trị trung bình P/E theo ngành cho cột 'pe_avg' của dataframe
    for index, co in df_dropColumns.iterrows():
        # 'co' is a tuple, so we can get the industry name like this
        industryName = co[11] #column index = 11
        # get the pe average value
        df_dropColumns.at[index, 'pe_avg'] =  df_pe_byIndustry.loc[industryName][0]
        
    df_filter_PE = df_dropColumns[df_dropColumns['priceToEarning'] < df_dropColumns['pe_avg']]
df_filter_PE
```

##### Lọc theo 3 tiêu chí: ROE >= 20% , EPS > 25000,  Book value per share > 10000

```python
df_filter_REB = df_filter_PE[(df_filter_PE['roe'] >= 0.17) & (df_filter_PE['earningPerShare'] > 2500) & (df_filter_PE['bookValuePerShare'] > 10000)]
df_filter_REB.reset_index(drop=True, inplace=True)
df_filter_REB
```

##### Lọc theo trung bình 20 phiên gần nhất lớn hơn 100.000

```python
# Tạo một df chứa ticker và trung bình Volume của 20 phiên gần nhất
df_volume = pd.DataFrame(columns=['ticker','volume_average', 'numSession'])
end_date = f'{trading_year}-07-01'
start_date = stfp.subtract_weekdays(end_date, 20)

with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=FutureWarning)
    for index, co in df_filter_REB.iterrows():
        # Tính tổng volume của 20 phiên gần nhất
        stockHisData = vnst.stock_historical_data(symbol=co['ticker'], start_date=start_date, end_date=end_date, resolution="1D", type="stock", beautify=True, decor=False, source='DNSE')        
        mean_volume = stockHisData['volume'].mean()
        stockHisData_len = len(stockHisData)
        df_volume.loc[len(df_volume)] = [co[0], mean_volume, stockHisData_len]
        print('Số phiên giao dịch lấy từ vnstock', co['ticker'], stockHisData_len)

# Thêm cột thông tin về trung bình Volume (volume_average) vào dataframe    
df_merged = pd.merge(df_filter_REB, df_volume[['ticker', 'volume_average']], on='ticker', how='left')

# Khối lượng giao dịch 20 phiên gần nhất > 100 ngàn
df_result = df_merged[df_merged['volume_average'] > 100000]
df_result.reset_index(drop=True, inplace=True)
```

```python
df_result.sort_values('priceToEarning', ascending=False)
df_result
```

#### Thông tin về 5 cổ phiếu lọc được

```python
filter_ticker_df = df_result.head(5)
filter_ticker_df
```

### Khởi tạo portfolio từ kết quả lọc cổ phiêu và tiến hành giao dịch

###### Khởi tạo portfolio


```python
ticker_list = filter_ticker_df['ticker'].to_list()
###### Thiết lập phần trăm danh mục dựa trên weight
df_combined =  wi.get_combined_df(ticker_list, trading_year)
weight_df = wi.cal_weight(df_combined, ticker_list)
initial_cash = 1000000000
my_portfolio = Portfolio(starting_cash=initial_cash, ticker_list=ticker_list, df_percentage=weight_df)
```

```python
my_portfolio.stock_df
```

#### Tạo dataframe tín hiệu

```python
signal_df = pd.DataFrame(columns=['index', 'time', 'open', 'high', 'low', 'close', 'volume', 'ticker',
                                    'on-balance_volume', 'OBV_label', 'close_t_minus_1', 'bar_type', 'para',
                                    'label_spread', 'close_bar_label', 'signal'])
calUtil = cal.DataProcessor
signal_df = cal.DataProcessor.get_signals_df(self=calUtil,ticker_list=ticker_list, trading_year=trading_year)
```

```python
signal_dropped_columns = signal_df[['time', 'ticker', 'open', 'high', 'low', 'close', 'volume', 'signal']]
signal_dropped_columns
```

#### Thực hiện xét tín hiệu để giao dịch

```python
date_performances_df = pd.DataFrame(columns=['date', 'performance']);
for date, group in signal_df.groupby('time'):
    for index, row in group.iterrows():
        my_portfolio.validate_transaction(signal_row=row)
    date_performance = my_portfolio.daily_performance()
    date_performances_df.loc[len(date_performances_df)] = [date, date_performance]
```

### Kết quả sau kì giao dịch

```python
portfolio_performance, per_per_tick = my_portfolio.portfolio_performance()
total_revenue = my_portfolio.calculate_holding_stock_values() + my_portfolio.cash_prop + my_portfolio.get_pending_money()
print("Số vốn ban đầu: ", '{:,.0f}'.format(initial_cash))
print("Hiệu suất toàn danh mục: ", round(portfolio_performance*100, 2), '%')
print("Tổng giá trị danh mục: ", '{:,.0f}'.format(total_revenue))
print('Tiền: ', '{:,.0f}'.format(my_portfolio.cash_prop))
print('Giá trị cổ phiếu đang nắm giữ: ', '{:,.0f}'.format(my_portfolio.calculate_holding_stock_values()))
print('Tiền chờ bán cổ phiếu: ', '{:,.0f}'.format(my_portfolio.get_pending_money()))
```
