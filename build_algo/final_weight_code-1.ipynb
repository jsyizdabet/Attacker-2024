```python
# Mức độ tương quan giữa các mã cp trong 20 phiên gần nhất (Rô)
corr_matrix = df_combined.corr()

# Calculate covariance matrix
def covariance_matrix(corr_matrix, volatility):
    n = len(volatility)
    cov_matrix = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            cov_matrix[i, j] = corr_matrix.iloc[i, j] * volatility[i] * volatility[j]  # Changed corr_matrix[i, j] to corr_matrix.iloc[i, j]
    return cov_matrix

# Calculate covariance matrix
cov_matrix = covariance_matrix(corr_matrix, volatility)

print("Covariance matrix:")
print(cov_matrix)
```

```python
# Bài toán cuối cùng
import numpy as np
from scipy.optimize import minimize

# Hàm mục tiêu: Minimize variance of portfolio
def objective(x, cov_matrix):
    return np.dot(x.T, np.dot(cov_matrix, x))

# Hàm điều kiện bằng 1
def constraint_eq(x):
    return sum(x) - 1

# Hàm điều kiện lớn hơn 0
def constraint_non_negative(x):
    return x

# Hàm điều kiện lớn hơn
def constraint_ge(x, threshold):
    return np.dot(threshold, x)

# Các ràng buộc
constraints = [
    {'type': 'eq', 'fun': constraint_eq},  # a + b + c + d + e = 1
    {'type': 'ineq', 'fun': lambda x: constraint_ge(x, df2['MR20'].values)}, #Lớn hơn 0
    {'type': 'ineq', 'fun': lambda x: constraint_ge(x, df3['MR90'].values) - 10},  # Lớn hơn 10
    {'type': 'ineq', 'fun': constraint_non_negative}  # a, b, c, d, e >= 0
]

# Giá trị khởi tạo
num_assets = cov_matrix.shape[0]  # Số lượng tài sản
x0 = np.ones(num_assets) / num_assets  # Trọng số khởi tạo bằng nhau

# Giải bài toán
result = minimize(objective, x0, args=(cov_matrix,), method='SLSQP', constraints=constraints)
# In kết quả thu được
result_df = pd.DataFrame({'weight': result['x'], 'objective value': result['fun']}, index=['a', 'b', 'c', 'd', 'e'])
result_df
```
